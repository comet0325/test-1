```python
data_forQ2= data_set[data_set['year'] == 2018]
```

```python
data_forQ2_C = data_forQ2[data_forQ2['cp'] == '포수']
data_forQ2_1B= data_forQ2[data_forQ2['cp'] == '1루수']
data_forQ2_2B = data_forQ2[data_forQ2['cp'] == '2루수']
data_forQ2_SS = data_forQ2[data_forQ2['cp'] == '유격수']
data_forQ2_3B = data_forQ2[data_forQ2['cp'] == '3루수']
data_forQ2_LF = data_forQ2[data_forQ2['cp'] == '좌익수']
data_forQ2_CF = data_forQ2[data_forQ2['cp'] == '중견수']
data_forQ2_RF = data_forQ2[data_forQ2['cp'] == '우익수']
```

```python
best_Catcher = data_forQ2_C.nlargest(1, 'war')
best_First = data_forQ2_1B.nlargest(1, 'war')
best_Second = data_forQ2_2B.nlargest(1, 'war')
best_ShortStop = data_forQ2_SS.nlargest(1, 'war')
best_Third = data_forQ2_3B.nlargest(1, 'war')
best_Left = data_forQ2_LF.nlargest(1, 'war')
best_Center = data_forQ2_CF.nlargest(1, 'war')
best_Right = data_forQ2_RF.nlargest(1, 'war')
print(best_Catcher)
print(best_First)
print(best_Second)
print(best_ShortStop)
print(best_Third)
print(best_Left)
print(best_Center)
print(best_Right)
```

```python
salary_R_Corr = data_set['salary'].corr(data_set['R'])
salary_H_Corr = data_set['salary'].corr(data_set['H'])
salary_HR_Corr = data_set['salary'].corr(data_set['HR'])
salary_RBI_Corr = data_set['salary'].corr(data_set['RBI'])
salary_SB_Corr = data_set['salary'].corr(data_set['SB'])
salary_war_Corr = data_set['salary'].corr(data_set['war'])
salary_avg_Corr = data_set['salary'].corr(data_set['avg'])
salary_OBP_Corr = data_set['salary'].corr(data_set['OBP'])
salary_SLG_Corr = data_set['salary'].corr(data_set['SLG'])
corr_List = {'stat': ['R', 'H', 'HR', 'RBI', 'SB', 'war', 'avg', 'OBP', 'SLG'], 
                'corr': [salary_R_Corr, salary_H_Corr, salary_HR_Corr, salary_RBI_Corr, salary_SB_Corr, 
                    salary_war_Corr, salary_avg_Corr, salary_OBP_Corr, salary_SLG_Corr]}
corr_DataFrame = pd.DataFrame(corr_List)
max_corr_idx = corr_DataFrame['corr'].idxmax()
print(corr_DataFrame.iloc[[max_corr_idx]])
```
